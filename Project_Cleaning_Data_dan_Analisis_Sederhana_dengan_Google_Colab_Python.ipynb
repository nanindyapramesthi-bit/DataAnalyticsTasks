# 2. Missing Values
Untuk handling missing values saya mengecek masing kolom pada statistika deskriptif diawal untuk mengetahui apakah pada setiap kolom tersebut ada outlier. Disini saya memutuskan untuk melakukan inputasi data untuk handling missing values yang proporsinya besar. Untuk name dan host name saya drop karena proporsi missing valuenya kecil. Untuk price saya isi dengan nilai median dikarenakan terdapat outlier. Untuk number of reviews saya isi dengan median dikarenakan terdapat outlier. Untuk last reviews saya isi dengan modus karena datanya berupa date time. Dan terakhir untuk reviews per month saya inputasi dengan median karena terdapat outlier.

```python
## Handling Missing Values
# Cek jumlah missing value
data.isnull().sum()
```

```python
# hitung proporsi missing value di tiap kolom
(data.isnull().sum()/5018)*100
```

```python
# Handling missing values
data.dropna(subset=['name', 'host_name'], inplace=True) # dihapus karena proporsi missing value terlalu kecil
data['price'].fillna(data['price'].median(), inplace=True)  # Isi dengan nilai median
data['number_of_reviews'].fillna(data['number_of_reviews'].median(), inplace=True)  # Isi dengan median
data['last_review'].fillna(data['last_review'].mode()[0], inplace=True)  # Isi dengan modus
data['reviews_per_month'].fillna(data['reviews_per_month'].median(), inplace=True)  # Isi dengan median
```

```python
# Menampilkan jumlah missing values untuk setiap kolom setelah handling
missing_values_after_handling = data.isnull().sum()
print(missing_values_after_handling)
```

# 4. Outliers
Disini saya mengecek jumlah outlier yang ada pada kolom price karena dilihat sekilas pada deskripsi statistika sebelumnya terdapat nilai maximum yang sangat jauh dari rata-rata. Langkah awal saya mengecek keseluruhan jumlah outlier untuk masing-masing kolom yang numeric, jika jumlahnya tidak terlalu banyak akan saya hapus outliernya tetapi jika jumlahnya banyak maka akan saya ganti outlier tersebut dengan nilai mediannya.

```python
# Handling Outliers
# Identifikasi outlier dengan boxplot atau metode lainnya
# untuk harga:
data.boxplot(column=['price'])
```

```python
# untuk minimum_nights:
data.boxplot(column=['minimum_nights'])
```

```python
# untuk number_of_reviews:
data.boxplot(column=['number_of_reviews'])
```

```python
# untuk reviews_per_month:
data.boxplot(column=['reviews_per_month'])
```

```python
# Menentukan Outlier price
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Count outliers
outliers_count = data[(data['price'] < lower_bound) | (data['price'] > upper_bound)].shape[0]
print("Jumlah nilai price outlier:", outliers_count)
```

```python
# Menentukan Outlier minimum_nights
Q1 = data['minimum_nights'].quantile(0.25)
Q3 = data['minimum_nights'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Count outliers
outliers_count = data[(data['minimum_nights'] < lower_bound) | (data['minimum_nights'] > upper_bound)].shape[0]
print("Jumlah nilai minimum_nights outlier:", outliers_count)
```

```python
# Menentukan Outlier number_of_reviews
Q1 = data['number_of_reviews'].quantile(0.25)
Q3 = data['number_of_reviews'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Count outliers
outliers_count = data[(data['number_of_reviews'] < lower_bound) | (data['number_of_reviews'] > upper_bound)].shape[0]
print("Jumlah nilai number_of_reviews outlier:", outliers_count)
```

```python
# Menentukan Outlier reviews_per_month
Q1 = data['reviews_per_month'].quantile(0.25)
Q3 = data['reviews_per_month'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Count outliers
outliers_count = data[(data['reviews_per_month'] < lower_bound) | (data['reviews_per_month'] > upper_bound)].shape[0]
print("Jumlah nilai reviews_per_month outlier:", outliers_count)
```

karena jumlah outlier banyak, sehingga jika dihapus akan kehilangan banyak informasi, saya memutuskan untuk mengganti nilai outlier dengan median

```python
data.shape
```

```python
# Hitung median untuk setiap kolom
median_price = data['price'].median()
median_min_nights = data['minimum_nights'].median()
median_num_reviews = data['number_of_reviews'].median()
median_reviews_per_month = data['reviews_per_month'].median()
```

```python
# Mengganti Outlier price
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data.loc[data['price'] > upper_bound, 'price'] = median_price
data.loc[data['price'] < lower_bound, 'price'] = median_price
# Count outliers
outliers_count = data[(data['price'] < lower_bound) | (data['price'] > upper_bound)].shape[0]
print("Jumlah nilai price outlier:", outliers_count)
```

```python
# Mengganti Outlier minimum_nights
Q1 = data['minimum_nights'].quantile(0.25)
Q3 = data['minimum_nights'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data.loc[data['minimum_nights'] > upper_bound, 'minimum_nights'] = median_min_nights
data.loc[data['minimum_nights'] < lower_bound, 'minimum_nights'] = median_min_nights
# Count outliers
outliers_count = data[(data['minimum_nights'] < lower_bound) | (data['minimum_nights'] > upper_bound)].shape[0]
print("Jumlah nilai minimum_nights outlier:", outliers_count)
```

```python
# Mengganti Outlier number_of_reviews
Q1 = data['number_of_reviews'].quantile(0.25)
Q3 = data['number_of_reviews'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data.loc[data['number_of_reviews'] > upper_bound, 'number_of_reviews'] = median_num_reviews
data.loc[data['number_of_reviews'] < lower_bound, 'number_of_reviews'] = median_num_reviews
# Count outliers
outliers_count = data[(data['number_of_reviews'] < lower_bound) | (data['number_of_reviews'] > upper_bound)].shape[0]
print("Jumlah nilai number_of_reviews outlier:", outliers_count)
```

```python
# Mengganti Outlier reviews_per_month
Q1 = data['reviews_per_month'].quantile(0.25)
Q3 = data['reviews_per_month'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data.loc[data['reviews_per_month'] > upper_bound, 'reviews_per_month'] = median_reviews_per_month
data.loc[data['reviews_per_month'] < lower_bound, 'reviews_per_month'] = median_reviews_per_month
# Count outliers
outliers_count = data[(data['reviews_per_month'] < lower_bound) | (data['reviews_per_month'] > upper_bound)].shape[0]
print("Jumlah nilai reviews_per_month outlier:", outliers_count)
```

Analisis Data

```python
summary_stats = data.describe()
print("Summary Statistics:")
summary_stats
```

```python
## Harga Rata-rata, Termurah, dan Termahal per Jenis Tipe Kamar
price_stats = data.groupby('room_type').agg({
    'price': ['mean', 'max', 'min'],
    'minimum_nights': lambda x: round(x.mean()), #pembulatan karena hari tidak mungkin desimal
    'number_of_reviews': lambda x: round(x.mean()) #pembulatan karena jumlah review bukan desimal
}).reset_index()
```

```python
print("\nAverage Price, Maximum Price, Minimum Price, and Other Metrics per Room Type:")
price_stats
```

```python
## Total Properti dan Rata-rata Harga per Neighbourhood Group
neighbourhood_stats = data.groupby('neighbourhood_group').agg({
    'id': 'count',
    'price': 'mean'
}).rename(columns={'id': 'total_properties', 'price': 'avg_price'})
```

```python
print("\nTotal Properties and Average Price per Neighbourhood Group:")
neighbourhood_stats
```

```python
## Rata-rata Review per Bulan dan Rata-rata Ketersediaan Kamar per Tahun
data['year'] = data['last_review'].dt.year
reviews_availability = data[(data['year'] >= 2017) & (data['year'] <= 2019)].groupby(['year', 'room_type']).agg({
    'reviews_per_month': 'mean',
    'availability_365': lambda x: round(x.mean())
}).reset_index()
```

```python
print("\nAverage Reviews per Month and Average Availability per Room Type per Year:")
reviews_availability
```

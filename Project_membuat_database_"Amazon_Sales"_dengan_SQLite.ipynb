```python
# Import library yang dibutuhkan
import pandas as pd
import sqlite3
```

Buatlah sebuah database dari file Excel yang diberikan dengan nama “amazon_sales.db”.

```python
# Membuat koneksi ke database SQLite
conn = sqlite3.connect("amazon_sales.db")
```

```python
# Membuat variabel cursor dari koneksi ke database
cursor = conn.cursor()
```

```python
# Cek database
query = """
SELECT name FROM sqlite_master WHERE type='table' AND name NOT LIKE 'sqlite_%';
"""
cursor.execute(query)
print(cursor.fetchall())
```

```python
#Membuat tabel amazonsales
query = """CREATE TABLE amazonsales (
  product_id VARCHAR NOT NULL,
  product_name VARCHAR NOT NULL,
  category VARCHAR NOT NULL,
  discounted_price INT NOT NULL,
  actual_price INT NOT NULL,
  discount_percentage FLOAT NOT NULL,
  qty INT NOT NULL,
  rating INT NOT NULL,
  rating_count INT NOT NULL,
  order_date DATE NOT NULL,
  arrival_date DATE NOT NULL,
  about_product VARCHAR NOT NULL,
  user_id VARCHAR NOT NULL,
  user_name VARCHAR NOT NULL,
  review_id VARCHAR NOT NULL,
  review_title VARCHAR NOT NULL,
  review_content VARCHAR NOT NULL
);"""

try:
    cursor.execute(query)
    print("Tabel berhasil dibuat!")
except Exception as e:
    print("Terjadi kesalahan")
    print(e)
```

```python
# Cek tabel amazonsales yang dibuat
query = """ PRAGMA table_info(amazonsales); """
pd.read_sql_query(query, conn)
```

No. 1

Sebutkan 5 kategori yang memiliki rata-rata jumlah potongan harga terbesar. Sebutkan juga masing-masing nilai dari kategori tersebut!

```python
query1 = """
SELECT category,
AVG(actual_price-discounted_price) AS avg_discount
FROM amazonsales
GROUP BY category
ORDER BY avg_discount DESC
LIMIT 5
"""
pd.read_sql_query(query1, conn)
```

No. 2

Berapa jumlah user yang memberikan review yang memuat kata 'Good Product' pada produk dengan kategori Computers & Accessories dan Office Products ?

```python
query2 = """
SELECT category,
COUNT(DISTINCT user_id) AS jumlah_review
FROM amazonsales
WHERE review_title LIKE '%Good Product%'
AND category IN ('Computers & Accessories', 'Office Products')
group by 1
"""
pd.read_sql_query(query2, conn)
```

No. 3

Tentukan 5 bulan yang memiliki total sales/revenue terendah pada tahun 2004 (dari seluruh kategori produk kecuali Office Products dan Home & Kitchen). Sebutkan juga nilai setiap revenue nya!

```python
query3 = """
SELECT strftime('%m', order_date) AS bulan, strftime('%Y', order_date) AS tahun,
       SUM(discounted_price * qty) AS total_sales
FROM amazonsales
WHERE category NOT IN ('Office Products', 'Home & Kitchen')
AND tahun = '2004'
GROUP BY bulan, tahun
ORDER BY total_sales
LIMIT 5
"""
pd.read_sql_query(query3, conn)
```

No. 4

Sebutkan kategori apa yang rata-rata waktu pengirimannya paling cepat dan lambat (Tanpa kategori Car & Motorbike dan Toys & Games).


```python
#Rata-rata pengiriman paling cepat
query4 = """
SELECT category,
date(arrival_date) AS arrival_date,
date(order_date) AS order_date,
round(AVG(strftime('%d',arrival_date)-strftime('%d',order_date))) AS lama_pengiriman
FROM amazonsales
WHERE category NOT IN('Car & Motorbike','Toys & Games')
GROUP BY category
ORDER BY lama_pengiriman ASC
limit 1
"""
pd.read_sql_query(query4,conn)
```

```python
#Rata-rata pengiriman paling lama
query4 = """
SELECT category,
date(arrival_date) AS arrival_date,
date(order_date) AS order_date,
round(AVG(strftime('%d',arrival_date)-strftime('%d',order_date))) AS lama_pengiriman
FROM amazonsales
WHERE category NOT IN('Car & Motorbike','Toys & Games')
GROUP BY category
ORDER BY lama_pengiriman DESC
limit 1
"""
pd.read_sql_query(query4,conn)
```

No. 5

Hitunglah AOV pada setiap tahunnya, namun data yang digunakan haruslah data dari awal hingga pertengahan bulan (tanggal 01 - 15) untuk setiap bulannya

```python
query5 = """
SELECT strftime('%Y', order_date) AS year,
       SUM(discounted_price*qty) AS total_sales,
       SUM(qty) AS total_order,
       SUM(discounted_price * qty) / SUM(qty) AS aov
FROM amazonsales
WHERE strftime('%d',order_date) BETWEEN '01' AND '15'
GROUP BY year
"""
pd.read_sql_query(query5, conn)
```

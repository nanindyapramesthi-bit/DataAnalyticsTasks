```python
# 1. Persentase growth pada total sales pada setiap bulannya
query_growth = """
SELECT
    periode,
    total_sales,
    LAG(total_sales) OVER (ORDER BY periode) AS prev_month_sales,
    ((total_sales - LAG(total_sales) OVER (ORDER BY periode)) / LAG(total_sales) OVER (ORDER BY periode)) * 100 AS sales_growth
FROM
    (SELECT
        SUBSTR(Order_Date, 1, 7) AS periode,
        SUM((product.Retail_Price * sales.Quantity) * (1 - customer.Loyalty_Discount)) AS total_sales
    FROM
        sales
    JOIN
        product ON sales.Item_id = product.Item_id
    JOIN
        customer ON sales.CustomerID = customer.CustomerID
    GROUP BY
        SUBSTR(Order_Date, 1, 7)
    ORDER BY
        periode) AS monthly_sales
"""
growth_df = pd.read_sql(query_growth, conn)
```

```python
# 2. Kategori pelanggan berdasarkan spending
query_category = """
WITH filtered_customers AS (
    SELECT
        c.CustomerID,
        CASE
            WHEN SUM(s.Quantity * p.Retail_Price) > 500 THEN 'High_spender'
            WHEN SUM(s.Quantity * p.Retail_Price) BETWEEN 300 AND 500 THEN 'Mid_spender'
            ELSE 'Low_spender'
        END AS cust_cat,
        SUM(s.Quantity) AS total_quantity,
        SUM(s.Quantity * p.Retail_Price) AS total_spending
    FROM sales s
    JOIN customer c ON s.CustomerID = c.CustomerID
    JOIN product p ON s.Item_id = p.Item_id
    WHERE s.Shipping_State NOT IN ('Wyomin', 'Idaho', 'Alaska')
    GROUP BY c.CustomerID
)
SELECT
    cust_cat,
    COUNT(DISTINCT CustomerID) AS total_user,
    CAST(AVG(total_quantity) AS INT)AS avg_quantity,
    AVG(total_spending) AS avg_spending
FROM filtered_customers
GROUP BY cust_cat
"""

category_df = pd.read_sql(query_category, conn)
```

```python
# 3. Ranking setiap produk pada tahun 2004 berdasarkan rata-rata rating
query_ranking = """
WITH product_rank AS (
    SELECT
        p.Item_id,
        AVG(s.Rating) AS avg_rating,
        ROW_NUMBER() OVER (ORDER BY AVG(s.Rating) DESC) AS ranking
    FROM  sales s
    JOIN  product p ON s.Item_id = p.Item_id
    WHERE SUBSTR(s.Order_Date, 1, 4) = '2004'
    GROUP BY p.Item_id
)
SELECT
    '2004' AS Year,
    p.Description AS Product,
    pr.avg_rating,
    pr.ranking
FROM product p
JOIN product_rank pr ON p.Item_id = pr.Item_id
WHERE pr.ranking <= 5
UNION ALL
SELECT
    '2004' AS Year,
    p.Description AS Product,
    pr.avg_rating,
    pr.ranking
FROM product p
JOIN product_rank pr ON p.Item_id = pr.Item_id
WHERE pr.ranking > (SELECT COUNT(*) FROM product_rank) - 5

"""

ranking_df = pd.read_sql(query_ranking, conn)
```
